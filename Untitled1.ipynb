```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate synthetic blobs dataset
# center=3 means 3 distinct blob groups will be generated
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=1.0, random_state=42)

# 2. Initialize and fit the K-Means algorithm to find 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# 3. Plotting the results using a scatter plot
plt.figure(figsize=(8, 6))
sns.set_theme(style="whitegrid")

# Create a scatter plot colored by the K-Means cluster labels
scatter = plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis', edgecolors='none')

# Add titles and axis labels to match standard visualization formats
plt.title("Figure 2: Synthetic Data Blob Visualization (Scikit-Learn blobs)", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)

# Generate a legend matching the colors to cluster IDs automatically
plt.legend(*scatter.legend_elements(), title="Clusters")

# Display the final plot
plt.show()
```
